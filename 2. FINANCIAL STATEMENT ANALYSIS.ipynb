```python
import numpy as np

revenue = np.array([14574.49, 7606.46, 8611.41, 9175.41, 8058.65, 8105.44, 11496.28, 9766.09, 10305.32, 14379.96, 10713.97, 15433.50])
expenses = np.array([12051.82, 5695.07, 12319.20, 12089.72, 8658.57, 840.20, 3285.73, 5821.12, 6976.93, 16618.61, 10054.37, 3803.96])
taxp_arr = np.array([])

for i in range(len(revenue)):
  profit = (revenue[i] - expenses[i])/1000
  profit_arr = np.array([])
  profit_arr = np.append(profit_arr,profit)

  print(f"profit: {profit:.2f}k")
  taxp = profit * 0.3
  taxp_arr = np.append(taxp_arr, taxp)
  profmarg = (taxp / revenue[i])*100*1000
  profmarg_arr = np.array([])
  profmarg_arr = np.append(profmarg_arr,profmarg)

  print(f"profit after tax: {taxp:.2f}k")
  print(f"profit margin: {profmarg:.2f}%")

  good_months = list([])
  bad_months = list([])

  for t in range(len(taxp_arr)):
    mean = sum(taxp_arr)/len(taxp_arr)
    if(taxp_arr[t] > mean):
      good_months.append(t)

    else:
      bad_months.append(t)


best_month = np.max(taxp_arr)
worst_month = np.min(taxp_arr)

print(f"\nGood months: {good_months}")
print(f"\nBad months: {bad_months}")

print(f"\nBest month: {best_month}")
print(f"\nWorst month: {worst_month}")
```
